```python
loans.info()
```

```python
##Creating a list of 1 element containing the string 'purpose'. Call this list cat_feats
```

```python
cat_feats = ['purpose']
```

```python
##Now we use pd.get_dummies(loans,columns=cat_feats,drop_first=True) to create a fixed larger dataframe 
##that has new feature columns with dummy variables. Setting this dataframe as final_data
```

```python
final_data = pd.get_dummies(loans,columns=cat_feats,drop_first=True)
```

```python
final_data.info()
```

```python
##Train Test Split
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = final_data.drop('not.fully.paid',axis=1)
y = final_data['not.fully.paid']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=101)
```

```python
## Training a decision tree model
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
##Creating an instance of DecisionTreeClassifier() called dtree and fit it to the training data.
```

```python
dtree = DecisionTreeClassifier()
```

```python
dtree.fit(X_train,y_train)
```

```python
##Creating predictions from the test set and creating a classification report and a confusion matrix
```

```python
predictions = dtree.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(y_test,predictions))
```

```python
print(confusion_matrix(y_test,predictions))
```
